```python
import numpy as np
import pandas as pd

# case = "Case1"
case = "DNN_pre"

results_dir = "../results/"
filename = results_dir + case + ".csv"
data = pd.read_csv(filename, header=None, index_col=None)
# data = data.drop(data.columns[0], axis=1)
data = data.iloc[1:,:]
# data
```

```python
data_np = data.to_numpy(dtype='float32')
nobs = data_np[:,0]
data_np = data_np[:,1:]
data_np = np.around(data_np,6)
# data_np
```
