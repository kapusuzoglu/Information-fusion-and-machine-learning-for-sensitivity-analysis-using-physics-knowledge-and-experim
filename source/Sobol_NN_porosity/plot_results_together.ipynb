```python
# case = ["DNN","DNN_loss","DNN_hyb","DNN_upd",
#         "DNN_loss_hyb","DNN_upd_hyb","DNN_upd_loss",
#         "DNN_upd_loss_hyb"]

# case = ["DNN_MC_first","DNN_loss_MC_first","DNN_hyb_MC_first","DNN_upd_MC_first",
#         "DNN_loss_hyb_MC_first","DNN_upd_hyb_MC_first","DNN_upd_loss_MC_first",
#         "DNN_upd_loss_hyb_MC_first","GP_first"]

case = ["DNN_MC_p_first_5000","DNN_loss_MC_p_first_5000","DNN_upd_MC_p_first_5000","DNN_upd_loss_MC_p_first_5000"]

# get the figure
fig = plt.figure()
ax = plt.subplot(111)
# labels = ['DNN$_{MC}$', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{hyb}}_{MC}$', 'DNN$^{\mathrm{upd}}_{MC}$', 'DNN$^{\mathrm{hyb}, \mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{upd, hyb}}_{MC}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{upd, hyb}, \mathcal{L}_{\mathrm{phy}}}_{MC}$','GP']
# labels = ['DNN$_{MC}$', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{upd}}_{MC}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}_{MC}$']
labels = ['DNN', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}$', 'DNN$^{\mathrm{upd}}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}$']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]

# Gp
results_dir = "../results/"

for j,i in enumerate(case):
    filename = results_dir + str(i) + ".csv"
    data = pd.read_csv(filename, header=None, index_col=None)
    data = data.iloc[1:,:]
    data_np = data.to_numpy(dtype='float32')
    nobs = data_np[:,0]
    data_np = data_np[:,1:]
    data_np = np.around(data_np,6)

    sobol = "S1_MC_first_5000"
    
    if (i == "DNN_hyb_MC_first" or i == "DNN_loss_hyb_MC_first" or i == "DNN_upd_hyb_MC_first" or i == "DNN_upd_loss_hyb_MC_first"):
        plt.plot(nobs, data_np[:,0], '-o', color=colors[j], label=labels[j])
        plt.plot(nobs, data_np[:,4], '_', color=colors[j])
        plt.plot(nobs, data_np[:,8], '_', color=colors[j])
    else:
        plt.plot(nobs, data_np[:,0], '-o', color=colors[j], label=labels[j])
        plt.plot(nobs, data_np[:,2], '_', color=colors[j])
        plt.plot(nobs, data_np[:,4], '_', color=colors[j])
                                                                                                                                              


    # plt.plot(obs[:,0], obs[:,1], 'o', color=colors[3])
    # plt.plot(pred_nn[:,0], pred_nn[:,1], '^', color=colors[0])
    # plt.plot(pred_nn_hyb[:,0], pred_nn_hyb[:,1], 'v', color=colors[1])
    # plt.plot(pred__piml[:,0], pred__piml[:,1], 'd', color=colors[4])
    # plt.plot(pred_piml_pre[:,0], pred_piml_pre[:,1], 's', color=colors[7])

plt.ylabel('$S_T$', fontsize=22)
plt.xlabel('n', fontsize=22)
plt.ylim((-0.01, 1))
plt.xlim((-0.01, 42))
# plt.xlim((-0.01, 1310))

plt.legend(loc='best', fontsize=12, fancybox=True)
# ax.legend(loc='right', bbox_to_anchor=(1.25, 0.5),
#           fancybox=True, shadow=False)
plt.tick_params(labelsize=20)
# ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.show()


plt_filename = '../figs/' +  sobol

# save as PDF
fig.savefig("{}.pdf".format(plt_filename), bbox_inches='tight', dpi=300)
```
