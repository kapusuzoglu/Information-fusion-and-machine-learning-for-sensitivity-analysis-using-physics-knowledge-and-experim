# Main effect for Temperature and Speed with MC Dropout

```python
# case = ["DNN_MC_p_Lake_first_10000","DNN_loss_MC_p_Lake_first_10000","DNN_upd_MC_p_Lake_first_10000","DNN_upd_loss_MC_p_Lake_first_10000"]
case = ["DNN_MC_p_Lake_first_5000","DNN_loss_MC_p_Lake_first_5000","DNN_upd_MC_p_Lake_first_5000","DNN_upd_loss_MC_p_Lake_first_5000"]

# case = ["DNN_MC_p_Lake_first_5000","DNN_loss_MC_p_Lake_first_5000"]


# labels = ['DNN$_{MC}$', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{hyb}}_{MC}$', 'DNN$^{\mathrm{upd}}_{MC}$', 'DNN$^{\mathrm{hyb}, \mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{upd, hyb}}_{MC}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{upd, hyb}, \mathcal{L}_{\mathrm{phy}}}_{MC}$','GP']
# labels = ['DNN$_{MC}$', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}_{MC}$', 'DNN$^{\mathrm{upd}}_{MC}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}_{MC}$']

labels = ['DNN', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}$', 'DNN$^{\mathrm{upd}}$', 'DNN$^{\mathrm{upd}, \mathcal{L}_{\mathrm{phy}}}$']

# labels = ['DNN', 'DNN$^{\mathcal{L}_{\mathrm{phy}}}$']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]

# Gp
results_dir = "../../../../figs/Lake/finalResults/"

# dimension
d = 11

for kk in range(d):
    # get the figure
    fig = plt.figure()
    ax = plt.subplot(111)
    sobol = "S" + str(kk+1) + "_MC_first_5000_Lake_edit"
    for j,i in enumerate(case):
        filename = results_dir + str(i) + ".csv"
        data = pd.read_csv(filename, header=None, index_col=None)
        data = data.iloc[1:6,:]
        data_np = data.to_numpy(dtype='float32')
        nobs = data_np[:,0]
        data_np = data_np[:,1:]
        data_np = np.around(data_np,6)
        
        plt.plot(nobs, data_np[:,kk], '-o', color=colors[j], label=labels[j])
        plt.plot(nobs, data_np[:,d+kk], '_', color=colors[j])
        plt.plot(nobs, data_np[:,2*d+kk], '_', color=colors[j])
                                                                                                                                              
        
    plt.ylabel('$X_{{{}}}$'.format(kk+1), fontsize=22)
    plt.xlabel('n', fontsize=22)
    plt.ylim((-0.01, 1))
#     plt.xlim((-0.01, 7500))
#     plt.xticks(np.arange(0, 13500, step=2500))
#     plt.xticks([500,2500,5000,7500])
    plt.xlim((100, 1000))
    plt.xticks([100,250,500,750,1000])

    plt.legend(loc='best', fontsize=12, fancybox=True)
    # ax.legend(loc='right', bbox_to_anchor=(1.25, 0.5),
    #           fancybox=True, shadow=False)
    plt.tick_params(labelsize=18)
    # ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    plt.show()

    plt_filename = '../../../../figs/Lake/finalResults/' +  sobol

    # save as PDF
    fig.savefig("{}.pdf".format(plt_filename), bbox_inches='tight', dpi=300)
```
